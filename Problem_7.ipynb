```python
# matplotlibのインポート
import matplotlib.pyplot as plt

# 紙の厚さ
THICKNESS = 0.00008

# 折られた紙の厚さ
folded_thickness = THICKNESS

# 紙を1回折った時の厚さを計算するコード
# folded_thickness = THICKNESS * 2

# 43回を指定
number_of_folding = 43

# 空のリストを作成+初期値の追加
folded_thickness_list = [folded_thickness]

# 折られた紙の厚さを求める計算式(for文での実行)
for number in range(number_of_folding):
	folded_thickness = folded_thickness * 2
	folded_thickness_list.append(folded_thickness)

# 格納数の確認
#print(len(folded_thickness_list))

# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list, color='red')
plt.show()
```

### 凡例(label)

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list, label='test')
plt.legend()
plt.show()
```

### 透過率(alpha)

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list, alpha=0.3)
plt.show()
```

### 線の種類を指定(linestyle)

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list, linestyle=':')
plt.show()
```

### 線の太さを指定(linewidth)

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list, linewidth=3)
plt.show()
```

### マーカー(marker)

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list, marker='o')
plt.show()
```

### マーカーの外部の色(markeredgecolor or mec)

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list, marker='o', markeredgecolor='red')
plt.show()
```

### マーカーの内部の色(markerfacecolor or mfc)

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list, marker='o', markerfacecolor ='red')
plt.show()
```

### マーカーのサイズ(markersize or ms)

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list, marker='o', markersize=10)
plt.show()
```

### マーカーの出現頻度(markevery)

- markevery = None(default)  
全てのマーカーの表示  
- markevery = 5  
5点おきにマーク  
- markevery = slice(0, 40, 5)  
スライスの書き方でも指定可能  
- maekevery = [2, 3, 5]  
指定番目のマークのみ表示

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list, marker='o', markevery=slice(0, 40, 5))
plt.show()
```
